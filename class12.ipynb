```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/Users/lin/Desktop/Study/Financial Big Data/ex9/clean_data.csv', sep = ',', index_col = 0)
price = data.iloc[:,1]
ret = np.log(price).diff()[2:]
```

```python
first_half = ret[:2305]
second_half = ret[2305:]
```

```python
l1_range = np.arange(1,200,3)
l2_range = np.arange(1,200,3)
performance = np.zeros((len(l1_range), len(l2_range)))
```

```python
l1 = 5
l2 = 10
```

```python
pos = np.sign(first_half.rolling(l1).sum() - first_half.rolling(l2).sum())
total_ret = pos[1:]*first_half[:-1]
```

```python
np.nansum(total_ret)
```

```python
i = 0
for l1 in l1_range:
    j = 0
    for l2 in l2_range:
        pos = np.sign(first_half.rolling(l1).sum() - first_half.rolling(l2).sum())
        total_ret = pos[1:]*first_half[:-1]
        performance[i,j] = np.nansum(total_ret)
        j = j+1
    i = i+1
```

```python
import seaborn as sns

sns.heatmap(performance)
```
